# Visualização de dados
> Introdução ao Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# fig, ax = plt.subplots(nrows=2,ncols=2)
```

```python
import pandas as pd
```

```python
# arq = "gapminder.csv"
arq="https://raw.githubusercontent.com/salgado/pos.lasalle.datascience/main/gapminder.csv"
df = pd.read_csv(arq)
```

```python
df.info()
```

```python
df.describe()
```

```python
# qual o pais mais pobre?
df[df['gdp']==142]
```

```python
# qual o pais mais rico?
df[df['gdp']==182668]
```

```python
# qual o pais com maior mortalidade infantil?
df[df['child_mortality']==435.8]
```

```python
df_brazil = df.query(" Country == 'Brazil' ").reset_index(drop=True)
df_brazil.describe()
```

```python
print(plt.rcParams.get('figure.figsize'))
```

```python
fig, ax = plt.subplots(figsize=[6,4], dpi=72)
ax.plot(df_brazil['Year'], df_brazil['gdp'])
plt.show()
```

```python
# Customizando gráfico

fig, ax = plt.subplots(figsize=[8,5], dpi=72)
ax.set_xlabel('Ano', fontdict={'size':12})
ax.set_ylabel('PIB', fontdict={'size':12})
ax.set_title("PIB do Brasil por Ano (1964-2013)")
ax.plot(df_brazil['Year'], df_brazil['gdp'])
plt.show()
```

```python
# Marcadores

fig, ax = plt.subplots(figsize=[9,6], dpi=72)
ax.set_xlabel('Ano', fontdict={'size':12})
ax.set_ylabel('PIB', fontdict={'size':12})
ax.set_title("PIB do Brasil por Ano (1964-2013)")
ax.plot(df_brazil['Year'], df_brazil['gdp'], marker='o')
plt.show()
```

```python
# Estilo da linha

fig, ax = plt.subplots(figsize=[9,6], dpi=72)
ax.set_xlabel('Ano', fontdict={'size':12})
ax.set_ylabel('PIB', fontdict={'size':12})
ax.set_title("PIB do Brasil por Ano (1964-2013)")
ax.plot(df_brazil['Year'], df_brazil['gdp'], linestyle='--', color='green')
plt.show()
```

```python
# Grid

fig, ax = plt.subplots(figsize=[9,6], dpi=72)
ax.set_xlabel('Ano', fontdict={'size':12})
ax.set_ylabel('PIB', fontdict={'size':12})
ax.set_title("PIB do Brasil por Ano (1964-2013)")
ax.plot(df_brazil['Year'], df_brazil['gdp'], linestyle='--', color='green')
plt.grid(True)
plt.show()
```

```python
# Definir intervalos do eixo X

fig, ax = plt.subplots(figsize=[12,8], dpi=72)
ax.set_xticks(list(range(df_brazil['Year'].min(), df_brazil['Year'].max()+1, 5)))
ax.set_xlabel('Ano', fontdict={'size':12})
ax.set_ylabel('PIB', fontdict={'size':12})
ax.set_title("PIB do Brasil por Ano (1964-2013)")
ax.plot(df_brazil['Year'], df_brazil['gdp'], linestyle='--', color='green')
plt.grid(True)
plt.show()
```

```python
# Legendas

fig, ax = plt.subplots(figsize=[12,8], dpi=72)
ax.plot(df_brazil['Year'], df_brazil['gdp'], linestyle='--', color='g', label="Brasil")

ax.set_xticks(list(range(df_brazil['Year'].min(), df_brazil['Year'].max()+1, 5)))
ax.set_xlabel('Ano', fontdict={'size':12})
ax.set_ylabel('PIB', fontdict={'size':12})
ax.set_title("PIB do Brasil por Ano (1964-2013)")
# ax.legend(loc='best')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, title="País", title_fontsize='large')

plt.grid(True)
plt.show()
```

```python
# Multiplos plots

fig, ax = plt.subplots(figsize=[12,8], dpi=72)

df_usa = df.query(" Country == 'United States' ").reset_index(drop=True)
df_argentina = df.query(" Country == 'Argentina' ").reset_index(drop=True)

ax.plot(df_argentina['Year'], df_argentina['gdp'], linestyle='--', color='b', label="Argentina")
ax.plot(df_brazil['Year'], df_brazil['gdp'], linestyle='-', color='g', label="Brasil")
ax.plot(df_usa['Year'], df_usa['gdp'], linestyle='--', color='r', label="EUA")

ax.set_xticks(list(range(df_brazil['Year'].min(), df_brazil['Year'].max()+1, 5)))
ax.set_xlabel('Ano', fontdict={'size':12})
ax.set_ylabel('PIB', fontdict={'size':12})
ax.set_title("PIB do Brasil por Ano (1964-2013)")
# ax.legend(loc='best')
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, title="País", title_fontsize='large')

plt.grid(True)
plt.show()
```

```python
# Dividir Plots

fig, ax = plt.subplots(nrows=3,ncols=1, figsize=[12,8], dpi=72)

df_usa = df.query(" Country == 'United States' ").reset_index(drop=True)
df_argentina = df.query(" Country == 'Argentina' ").reset_index(drop=True)

ax[0].plot(df_argentina['Year'], df_argentina['gdp'], linestyle='--', color='b', label="Argentina")
ax[1].plot(df_brazil['Year'], df_brazil['gdp'], linestyle='-', color='g', label="Brasil")
ax[2].plot(df_usa['Year'], df_usa['gdp'], linestyle='--', color='r', label="EUA")

plots = [0,1,2]
for x in plots :
    ax[x].set_xticks(list(range(df_brazil['Year'].min(), df_brazil['Year'].max()+1, 5)))
    ax[x].set_xlabel('Ano', fontdict={'size':12})
    ax[x].set_ylabel('PIB', fontdict={'size':12})
    handles, labels = ax[x].get_legend_handles_labels()
    ax[x].legend(handles, labels, title="País", title_fontsize='large')
    ax[x].grid(True)

plt.show()
```

```python
# Twinx

fig, ax = plt.subplots(figsize=[12,8], dpi=72)

df_usa = df.query(" Country == 'United States' ").reset_index(drop=True)
df_argentina = df.query(" Country == 'Argentina' ").reset_index(drop=True)

ax.plot(df_brazil['Year'], df_brazil['gdp'], linestyle='-', color='g', label="Brasil")
ax.set_xticks(list(range(df_brazil['Year'].min(), df_brazil['Year'].max()+1, 5)))
ax.set_xlabel('Ano', fontdict={'size':12})
ax.set_ylabel('PIB Brasil', fontdict={'size':12})
ax.set_title("PIB do Brasil por Ano (1964-2013)")

ax2 = ax.twinx()
ax2.plot(df_usa['Year'], df_usa['gdp'], linestyle='--', color='r', label="EUA")
ax2.set_ylabel('PIB EUA', fontdict={'size':12})

handles, labels = ax.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax.legend(handles + handles2, labels + labels2, title="País", title_fontsize='large')

plt.grid(True)
plt.show()
```

```python
# Anotações

fig, ax = plt.subplots(figsize=[12,8], dpi=72)

df_china = df.query(" Country == 'China' ").reset_index(drop=True)

ax.plot(df_china['Year'], df_china['fertility'], linestyle="-", color="g", label='Brasil')

ax.set_xlabel('Ano', fontdict={'size': 12})
ax.set_ylabel('Fertilidade', fontdict={'size': 12})
ax.set_title('Fertilidade China Evolução (1964-2013)')

ax.annotate("Política de 2 filhos", xy=[1967,5.25], xytext=(1963.35,4.72), arrowprops={'arrowstyle':"->"})
ax.annotate("Política de 1 filho", xy=[1980,2.30], xytext=(1976.55,1.78), arrowprops={'arrowstyle':"->"})

plt.grid(True)
plt.show()
```
